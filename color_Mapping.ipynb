```python
import random 
```

# Reading File from the folder

```python
nodes=[]
with open('graph.txt') as f:
    lines=f.readlines()
for i in range(len(lines)):
    nodes.append(lines[i].split())
graph={}
for i in range(len(nodes)):
    for j in range(len(nodes[i])):
        key=nodes[i][0]
        graph[key]= nodes[i][1:j+1]

             
```

# After reading the file Our graph looks like this

```python
print(graph)
```

# CSP part 

```python
# Assume forward cheking 
colors=['Red', 'Green', 'Blue', 'Yellow', 'Violet', 'Gray', 'Orange']
color_map={} # A data structure that maps our data from the graph to color assignement 
pickColor=random.choice(colors) # Pick Random color from the list and assign for a given node if it satisfied CSP

for key in graph.keys():
    pickColor=random.choice(colors)
     # check if there are neighbours and compare their color with the node 
        #if they are the same back and pick another color again
    if len(graph[key])>0: 
        for i in range(len(graph[key])):
            if not pickColor in color_map.values():
                color_map[key]=pickColor
        print('Colors were enough!')
    else:
        print("Not enough colors!")
print(color_map,end="\n")
```
